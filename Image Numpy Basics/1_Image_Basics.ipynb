#### Image and Numpy Basics

In this section we will discuss image basics applying Numpy arrays to images

---

How does a computer represent image data?

 - Each image can be represented as an array.
 - Ex. a number is presented by pixels (such as 28 $\times$ 28 gray scale).
 - More resolution (more pixels)
 - 28 $\times$ 28 pixels are represented by arrays of [0,1] and these numbers can decided how dark or light an image can be presented.
 - Default images have values between 0-255 (255 is the lighest value & 0 is the darkest)
 - [0-255] is chosen according to how the computers store 8 bit numbers
 - <b> We always normalize [0-255] values to have between [0-1]</b>

Color images:
 - Color images can be represented as a combination of Red, Green and Blue (RGB) that allows to create different range of colors.
 - Each color channel will have intensity values.
 - In python, shape of color array has 3 diimensions <b>(Height, Width, Color Channels)</b>
 - Example - (1270,720,3): 3D array , one for each color
  - pixel width = 1270
  - pixel height = 720
  - color channels = 3
 - user needs to dictate which color is for which color as computer read every channle as gray scale [0-255].

Numpy can't open directly jpeg file so we need to use PIL <b> (Python Imaging Library )</b>

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from PIL import Image
```

```python
pic = Image.open('00-puppy.jpg')
pic
```

```python
type(pic)
```

So this Jpeg file so NumPy can't function on such extension files so we need to cerate the image into array

```python
pic_arr = np.asarray(pic)
type(pic_arr)
```

```python
pic_arr.shape
```

So this image has <b> 1300 pixel width, 1950 pixel height and 3 color channels </b>

```python
plt.imshow(pic_arr)
```

```python
# Accessing one of the channel

pic_red = pic_arr.copy()
pic_green = pic_arr.copy()
pic_blue = pic_arr.copy()

#zero out green and blue channel
# R G B - 0 1 2

plt.imshow(pic_red[:,:,0])
```

However the iamge above for red doesn't show red image has function between [0-255] but matplotlib function will show image by default like this with color scale

```python
plt.imshow(pic_red[:,:,0],cmap='gray') # in gray scale

# red channel - 0 = no red, 255 - pure red
# in gray scale 0 - black, 255 - pure white (more red)
```

```python
plt.imshow(pic_red[:,:,1],cmap='gray') #in green channel
```

```python
plt.imshow(pic_red[:,:,2],cmap='gray') #in blue channel
```

```python
#how to see red channel only without matplotlib default coloring function

pic_red[:,:,1]=0 #green channel is zero
pic_red[:,:,2]=0 # blue channel is zero
plt.imshow(pic_red)
```

```python
pic_green[:,:,0]=0 # red channel is zero
pic_green[:,:,2]=0 # blue channel is zero
plt.imshow(pic_green)
```

```python
pic_blue[:,:,0]=0 # red channel is zero
pic_blue[:,:,1]=0 # green channel is zero
plt.imshow(pic_blue)
```
