##### 1. Color Spaces

 - We worked with RGB which is very commonly used. In 1970s, <b> HSL (hue,saturation, lightness) and HSV (hue, saturation, value)</b> color models were used.
 - HSL/HSV are more closely aligned with the way human vision actually perceives color.

How HSL/HSV are different from RGB

- RGB can be though of 3D cube with three dimensions of Red, Green and Blue.
- HSL can be thought as a cylinder model where:
  - <b> Hue: around the cylinder different colors
  - Saturation: how much huw you end up having
  - Lightness: how dark (bottom of cylinder - black) or light (top of cylinder - white) </b>
- HSV can be also thought as a cylinder model like HSL where:
  - <b> Hue: around the cylinder different colors
  - Saturation: how much hue you end up having
  - Value: white is at the top center only

Conversion from RGB to HSV

```python
R = int(input("Enter R value: "))
G = int(input("Enter G value: "))
B = int(input("Enter B value: "))
```

0 to 255: one digit: 0 to 9
two digt: 10 to 99
three digit: 100 to 255

```python
# Constraining the values to the range 0 to 1: Normalizing it to 0 to 1

R_dash = R / 255
G_dash = G / 255
B_dash = B / 255

# defining the following terms for convenience: delta is range

Cmax = max(R_dash, G_dash, B_dash)
Cmin = min(R_dash, G_dash, B_dash)
delta = Cmax - Cmin

# hue calculation
if (delta == 0):
  H = 0
elif (Cmax == R_dash):
  H = (60 * (((G_dash  - B_dash) / delta) % 6))
elif (Cmax == G_dash):
  H = (60 * (((B_dash  - R_dash) / delta) + 2))
elif (Cmax == B_dash):
  H = (60 * (((R_dash  - G_dash) / delta) + 4))
 
# saturation calculation
if (Cmax == 0):
  S = 0
else:
  S = delta / Cmax

# value calculation
V = Cmax

# print output. H in degrees. S and V in percentage. 
# these values may also be represented from 0 to 255.
print("H = {:.1f}".format(H))
print("S = {:.1f}%".format(S * 100))
print("V = {:.1f}%".format(V * 100))
```
